```python
tp = 0
tn = 0
fp = 0
fn = 0

for i in range(len(feature_matrix)):
    score = df_features.iloc[i]["score"]
    label = df_features.iloc[i]["label"]

    if score >= 0.5:
        if label == "IndexCancer":
            tp += 1
        else:
            fp += 1
    else:
        if label == "IndexCancer":
            fn += 1
        else:
            tn += 1
```

```python
import seaborn as sns
import pandas as pd

# Assuming tp, tn, fp, fn are already calculated
confusion_matrix = [[tp, fp],
                    [fn, tn]]

# Convert to DataFrame for labeling axes
df_cm = pd.DataFrame(confusion_matrix, index=["Actual Positive", "Actual Negative"], 
                     columns=["Predicted Positive", "Predicted Negative"])

# Plot
plt.figure(figsize=(10,7))
sns.heatmap(df_cm, annot=True, cmap="Blues", fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
